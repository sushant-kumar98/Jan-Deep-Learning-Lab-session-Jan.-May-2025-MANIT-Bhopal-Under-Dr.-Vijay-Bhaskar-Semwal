```python
import numpy as np
import matplotlib.pyplot as plt

f1 = lambda x: -3 * x**2 + 2 * x
f2 = lambda x: x**3 + 2 * x**2 - 5
x = np.linspace(-5, 5, 500)

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(x, f1(x), label='-3x^2 + 2x', color='blue')
plt.title('Function: -3x^2 + 2x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, f2(x), label='x^3 + 2x^2 - 5', color='red')
plt.title('Function: x^3 + 2x^2 - 5')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()

plt.tight_layout()
plt.show()

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z1 = -3 * x**2 + 2 * y
z2 = x**3 + 2 * y**2 - 5

plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
contour1 = plt.contour(x, y, z1, levels=50, cmap='viridis')
plt.title('Contour: -3x^2 + 2y')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(contour1)
754338
plt.subplot(1, 2, 2)
contour2 = plt.contour(x, y, z2, levels=50, cmap='plasma')
plt.title('Contour: x^3 + 2y^2 - 5')
plt.xlabel('x')
plt.ylabel('y')
plt.colorbar(contour2)

plt.tight_layout()
plt.show()
```
