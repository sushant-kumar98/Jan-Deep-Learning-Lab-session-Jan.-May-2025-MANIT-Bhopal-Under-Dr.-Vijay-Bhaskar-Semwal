```python
import numpy as np
import matplotlib.pyplot as plt

# Define functions and their gradients
def f(x):
    return x**2

def df(x):
    return 2*x

def G(x):
    return x**4 + x**3 + x**2

def dG(x):
    return 4*x**3 + 3*x**2 + 2*x

# Gradient Descent Variants
def gradient_descent(grad, x_init, lr=0.01, epochs=50, method="batch", batch_size=10):
    np.random.seed(42)
    x = x_init
    trajectory = [x]

    for epoch in range(epochs):
        if method == "batch":
            x -= lr * grad(x)  # Full batch update
        elif method == "sgd":
            x_sample = x + np.random.randn()  # Simulate random sampling
            x -= lr * grad(x_sample)
        elif method == "mini-batch":
            x_batch = x + np.random.randn(batch_size)  # Simulate mini-batch
            x -= lr * np.mean(grad(x_batch))

        trajectory.append(x)

    return np.array(trajectory)

# Initialize parameters
np.random.seed(42)
x_init = np.random.uniform(-20, 20)
epochs = 50
lr = 0.01

# Run gradient descent for f(x) = x^2
traj_f_gd = gradient_descent(df, x_init, lr, epochs, "batch")
traj_f_sgd = gradient_descent(df, x_init, lr, epochs, "sgd")
traj_f_mbgd = gradient_descent(df, x_init, lr, epochs, "mini-batch")

# Run gradient descent for G(x) = x^4 + x^3 + x^2
traj_G_gd = gradient_descent(dG, x_init, lr, epochs, "batch")
traj_G_sgd = gradient_descent(dG, x_init, lr, epochs, "sgd")
traj_G_mbgd = gradient_descent(dG, x_init, lr, epochs, "mini-batch")

# Plot trajectories for f(x)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(traj_f_gd, label="Batch GD")
plt.plot(traj_f_sgd, label="SGD")
plt.plot(traj_f_mbgd, label="Mini-batch SGD")
plt.xlabel("Epochs")
plt.ylabel("x value")
plt.title("Convergence for f(x) = x^2")
plt.legend()

# Plot trajectories for G(x)
plt.subplot(1, 2, 2)
plt.plot(traj_G_gd, label="Batch GD")
plt.plot(traj_G_sgd, label="SGD")
plt.plot(traj_G_mbgd, label="Mini-batch SGD")
plt.xlabel("Epochs")
plt.ylabel("x value")
plt.title("Convergence for G(x) = x^4 + x^3 + x^2")
plt.legend()

plt.show()

# Final values of x after optimization
x_f_gd, x_f_sgd, x_f_mbgd = traj_f_gd[-1], traj_f_sgd[-1], traj_f_mbgd[-1]
x_G_gd, x_G_sgd, x_G_mbgd = traj_G_gd[-1], traj_G_sgd[-1], traj_G_mbgd[-1]

print("Final optimized x values:")
print(f"f(x) = x^2 -> Batch GD: {x_f_gd}, SGD: {x_f_sgd}, Mini-batch SGD: {x_f_mbgd}")
print(f"G(x) = x^4 + x^3 + x^2 -> Batch GD: {x_G_gd}, SGD: {x_G_sgd}, Mini-batch SGD: {x_G_mbgd}")
```

```python
# Define the new function and its derivative
def cos_f(x):
    return np.cos(np.pi * x)

def d_cos_f(x):
    return -np.pi * np.sin(np.pi * x)  # Corrected derivative

# Run gradient descent for cos(pi*x) in range x = [-4, 4]
x_init = np.random.uniform(-4, 4)

traj_cos_gd = gradient_descent(d_cos_f, x_init, lr, epochs, "batch")
traj_cos_sgd = gradient_descent(d_cos_f, x_init, lr, epochs, "sgd")
traj_cos_mbgd = gradient_descent(d_cos_f, x_init, lr, epochs, "mini-batch")

# Plot convergence trajectory for cos(pi*x)
plt.figure(figsize=(6, 5))
plt.plot(traj_cos_gd, label="Batch GD")
plt.plot(traj_cos_sgd, label="SGD")
plt.plot(traj_cos_mbgd, label="Mini-batch SGD")
plt.xlabel("Epochs")
plt.ylabel("x value")
plt.title("Convergence for f(x) = cos(pi*x)")
plt.legend()
plt.show()

# Final values of x after optimization for cos(pi*x)
x_cos_gd, x_cos_sgd, x_cos_mbgd = traj_cos_gd[-1], traj_cos_sgd[-1], traj_cos_mbgd[-1]

print(f"\n\nf(x) = Cos(pi.x) -> Batch GD: {x_cos_gd}, SGD: {x_cos_sgd}, Mini-batch SGD: {x_cos_mbgd}")
```

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Step 1: Load Dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ["Pregnancies", "Glucose", "BloodPressure", "SkinThickness", "Insulin",
           "BMI", "DiabetesPedigreeFunction", "Age", "Outcome"]
df = pd.read_csv(url, names=columns)

# Step 2: Data Exploration
print(df.head())  # Display first few rows
print(df.describe())  # Summary statistics
print(df.isnull().sum())  # Check for missing values

# Step 3: Handling Missing Values (Replace 0s with NaN in certain columns)
zero_replace_cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[zero_replace_cols] = df[zero_replace_cols].replace(0, np.nan)

# Impute missing values with median
df.fillna(df.median(), inplace=True)

# Step 4: Handle Class Imbalance (Optional: Use Class Weights)
class_weights = {0: 1, 1: df["Outcome"].value_counts()[0] / df["Outcome"].value_counts()[1]}

# Step 5: Splitting Data
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# Step 6: Feature Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Step 7: Build Neural Network Model
model = Sequential([
    Dense(16, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.3),  # Prevent Overfitting
    Dense(8, activation="relu"),
    Dense(1, activation="sigmoid")  # Output layer for binary classification
])

model.compile(optimizer=Adam(learning_rate=0.001),
              loss="binary_crossentropy",
              metrics=["accuracy"])

# Step 8: Train the Model
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test), class_weight=class_weights)

# Step 9: Evaluate the Model
y_pred_prob = model.predict(X_test)
y_pred = (y_pred_prob > 0.5).astype(int)

# Classification Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-score: {f1:.2f}")
print(f"ROC-AUC: {roc_auc:.2f}")

# Step 10: Plot Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,4))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# Step 11: ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_pred_prob)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle="--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()

# Step 12: Plot Training Curves
plt.figure(figsize=(10, 4))

# Loss Curve
plt.subplot(1, 2, 1)
plt.plot(history.history["loss"], label="Train Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve")
plt.legend()

# Accuracy Curve
plt.subplot(1, 2, 2)
plt.plot(history.history["accuracy"], label="Train Accuracy")
plt.plot(history.history["val_accuracy"], label="Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.title("Accuracy Curve")
plt.legend()

plt.show()
```
